```python
%%sql

WITH province_totals AS (-- This CTE calculates the population of each province
    SELECT
        province_name,
        SUM(number_of_people_served) AS total_people_served
    FROM
        combined_analysis_table
    GROUP BY
        province_name
)
SELECT
    ct.province_name,
    -- These case statements create columns for each type of source.
    -- The results are aggregated and percentages are calculated
    ROUND((SUM(CASE WHEN type_of_water_source = 'river'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / pt.total_people_served), 0) AS river,
    ROUND((SUM(CASE WHEN type_of_water_source = 'shared_tap'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / pt.total_people_served), 0) AS shared_tap,
    ROUND((SUM(CASE WHEN type_of_water_source = 'tap_in_home'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / pt.total_people_served), 0) AS tap_in_home,
    ROUND((SUM(CASE WHEN type_of_water_source = 'tap_in_home_broken'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / pt.total_people_served), 0) AS tap_in_home_broken,
    ROUND((SUM(CASE WHEN type_of_water_source = 'well'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / pt.total_people_served), 0) AS well
FROM
    combined_analysis_table ct
JOIN
    province_totals pt ON ct.province_name = pt.province_name
GROUP BY
    ct.province_name
ORDER BY
    ct.province_name;
```

## Key Patterns & Priorities
#### 1. **Sokoto** — Heavy river dependence

-- River users: 21 (highest of all provinces)

-- Immediate priority for well drilling and filtration to reduce unsafe river water reliance.

-- A well project here would directly help the largest at-risk group.

#### 2. **Amanzi** — Tap majority but infrastructure failure

-- Tap in home: 28, Tap in home broken: 24 → ~46% of taps are non-functional.

-- Shared tap also high (38), suggesting strong piped-water network presence — just poorly maintained.

-- Priority: Repair infrastructure (pumps, treatment plants, reservoirs) to restore water supply quickly to thousands.

#### 3. **Kilimani** — High shared tap usage

-- Shared tap: 47 (largest in dataset), but in-home taps are low (13) with 12 broken.

-- Could be a sign of urban areas relying on community points rather than individual connections.

-- Priority: Improve shared tap reliability or expand in-home connections.

#### 4. **Akatsi** & **Hawassa** — Similar patterns

-- Both have moderate river use (5 & 4) but rely heavily on shared taps (49 & 43).

-- Wells are significant (23 & 24), meaning groundwater resources are already in use — potential for more well expansion to reduce shared tap dependency.

### Other Notable Observations
-- ***High tap breakage is common***: Amanzi (24), Hawassa (15), Kilimani (12) — repairing these may be faster and cheaper than drilling new wells.

-- ***Wells are already widely used***: All provinces have wells, meaning groundwater is accessible — making well expansion a viable option in most areas.

-- ***Shared tap reliance dominates***: Every province’s shared tap numbers are high (38–49), meaning improvements to these systems would have wide impact.

```python
%%sql

WITH town_totals AS ( -- This CTE calculates the population for each province/town pair
    SELECT
        province_name,
        town_name,
        SUM(number_of_people_served) AS total_people_served
    FROM
        combined_analysis_table
    GROUP BY
        province_name, town_name
)
SELECT
    ct.province_name,
    ct.town_name,
    ROUND((SUM(CASE WHEN type_of_water_source = 'river'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / tt.total_people_served), 0) AS river,
    ROUND((SUM(CASE WHEN type_of_water_source = 'shared_tap'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / tt.total_people_served), 0) AS shared_tap,
    ROUND((SUM(CASE WHEN type_of_water_source = 'tap_in_home'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / tt.total_people_served), 0) AS tap_in_home,
    ROUND((SUM(CASE WHEN type_of_water_source = 'tap_in_home_broken'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / tt.total_people_served), 0) AS tap_in_home_broken,
    ROUND((SUM(CASE WHEN type_of_water_source = 'well'
                    THEN number_of_people_served ELSE 0 END) * 100.0 / tt.total_people_served), 0) AS well
FROM
    combined_analysis_table ct
JOIN
    town_totals tt 
    ON ct.province_name = tt.province_name
    AND ct.town_name = tt.town_name
GROUP BY
    ct.province_name, ct.town_name
ORDER BY
    ct.province_name, ct.town_name;
```

## Insights

A couple of weeks ago, we identified the following key insights:

1. **Most water sources are rural** in Maji Ndogo.  
2. **43%** of the population uses shared taps — with as many as **2,000 people sharing a single tap**.  
3. **31%** of the population has water infrastructure in their homes.  
4. Of those with home infrastructure, **45%** face **non-functional systems** due to broken pipes, pumps, and reservoirs.  
   - Affected areas include Amina, rural parts of Amanzi, and several towns in Akatsi and Hawassa.  
5. **18%** of the population uses wells — but only **28%** of these wells provide clean water.  
   - Contaminated wells are mostly found in Hawassa, Kilimani, and Akatsi.  
6. **Average wait times** for water are over **120 minutes**.  
   - Queues are longest on **Saturdays**.  
   - Queues are longer in the **mornings and evenings**.  
   - **Wednesdays and Sundays** have the shortest queues.


## Plan of Action

1. **Focus on high-impact improvements**  
   - Prioritize improving **shared taps**, as they serve the largest number of people.

2. **Improve water quality from wells**  
   - Wells are a good water source, but many are **contaminated**. Cleaning and treating them will benefit many communities.

3. **Repair existing infrastructure**  
   - Restoring running water in homes will eliminate the need for many to queue, which also shortens wait times for others — solving **two problems at once**.

4. **Avoid large-scale new tap installations for now**  
   - Installing taps in every home will **stretch resources too thin**. Only improve this source if **queue times are high**.

5. **Prepare for rural repair challenges**  
   - Most water sources are in **rural areas**. Teams should plan for difficult road access, supply shortages, and labor constraints.

## Practical Solutions

1. **Communities using rivers**  
   - **Short-term:** Dispatch trucks to provide water temporarily.  
   - **Long-term:** Send crews to drill wells for a permanent solution.  
   - **First target:** Sokoto province.

2. **Communities using wells**  
   - Install filters to purify water.  
   - **Chemically polluted wells:** Install reverse osmosis (RO) filters.  
   - **Biologically contaminated wells:** Install UV filters (alongside RO filters).  
   - **Long-term:** Investigate causes of pollution to prevent recurrence.

3. **Shared taps**  
   - **Short-term:** Send additional water tankers to the busiest taps on the busiest days (using the queue time pivot table to schedule deliveries).  
   - **Medium-term:** Install additional taps to reduce queue times to below **30 minutes** (UN standard).  
   - **Priority towns:** Bello, Abidjan, and Zuri.

4. **Shared taps with short queue times (< 30 min)**  
   - Further reduction is logistically challenging.  
   - Installing taps in homes is the most effective solution but is resource-intensive — better as a **long-term goal**.

5. **Broken infrastructure**  
   - High impact even from a single repair (e.g., fixing a reservoir or main pipe that serves many taps).  
   - **Priority areas:** Amina, Lusaka, Zuri, Djenne, and rural parts of Amanzi.

```python
%%sql

DROP TABLE IF EXISTS Project_progress;
CREATE TABLE Project_progress (
    Project_id SERIAL PRIMARY KEY,
    /* Project_id – Unique key for sources in case we visit 
       the same source more than once in the future. */
    source_id VARCHAR(20) NOT NULL 
        REFERENCES water_source(source_id) 
        ON DELETE CASCADE 
        ON UPDATE CASCADE,
    /* source_id – Each of the sources we want to improve should exist,
       and should refer to the water_source table. This ensures data integrity. */
    Address VARCHAR(50), -- Street address
    Town VARCHAR(30),
    Province VARCHAR(30),
    Source_type VARCHAR(50),
    Improvement VARCHAR(50), -- What the engineers should do at that place
    Source_status VARCHAR(50) DEFAULT 'Backlog' 
        CHECK (Source_status IN ('Backlog', 'In progress', 'Complete')),
    /* Source_status – We want to limit the type of information engineers can give us.
       - By DEFAULT all projects are in the "Backlog", which is like a TODO list.
       - CHECK() ensures only those three options will be accepted.
       This helps to maintain clean data. */
    Date_of_completion DATE, -- Engineers will add this the day the source has been upgraded.
    Comments TEXT -- Engineers can leave comments (TEXT has no char length limit).
);

INSERT INTO Project_progress (
    Source_id,
    Address,
    Town,
    Province,
    Source_type,
    Comments,
    Improvement
)
SELECT
    water_source.source_id AS Source_id,
    location.address AS Address,
    location.town_name AS Town,
    location.province_name AS Province,
    water_source.type_of_water_source AS Source_type,
    well_pollution.results AS Comments,
    CASE
        -- Step 1: Wells
        WHEN water_source.type_of_water_source = 'well'
             AND LOWER(well_pollution.results) = 'Contaminated: Chemical'
        THEN 'Install RO filter'

        WHEN water_source.type_of_water_source = 'well'
             AND LOWER(well_pollution.results) = 'Contaminated: Biological'
        THEN 'Install UV filter'

        -- Step 2: Rivers
        WHEN water_source.type_of_water_source = 'river'
        THEN 'Drill well'

        -- Step 3: Shared taps (queue time >= 30 min)
        WHEN water_source.type_of_water_source = 'shared_tap'
             AND visits.time_in_queue >= 30
        THEN CONCAT('Install ', FLOOR(visits.time_in_queue / 30), ' taps nearby')

        -- Step 4: Broken in-home taps
        WHEN water_source.type_of_water_source = 'tap_in_home_broken'
        THEN 'Diagnose local infrastructure'

        -- No NULLs allowed, so just in case
        ELSE 'No improvement needed'
    END AS Improvement
FROM
    water_source
LEFT JOIN
    well_pollution ON water_source.source_id = well_pollution.source_id
INNER JOIN
    visits ON water_source.source_id = visits.source_id
INNER JOIN
    location ON location.location_id = visits.location_id
WHERE
    visits.visit_count = 1
    AND (
        (water_source.type_of_water_source = 'well'
            AND well_pollution.results IS NOT NULL
            AND LOWER(well_pollution.results) <> 'clean')
        OR water_source.type_of_water_source IN ('tap_in_home_broken', 'river')
        OR (water_source.type_of_water_source = 'shared_tap'
            AND visits.time_in_queue >= 30)
    );
```
